```python
#Change'Followers', 'Authentic engagement\r\n' and 'Engagement avg\r\n' values to float

def value_to_float(x):
    if type(x) == float or type(x) == int:
        return x
    if 'K' in x:
        if len(x) > 1:
            return float(x.replace('K', '')) * 1000
        return 1000.0
    if 'M' in x:
        if len(x) > 1:
            return float(x.replace('M', '')) * 1000000
        return 1000000.0
    if 'B' in x:
        return float(x.replace('B', '')) * 1000000000
    return 0.0

instagram_data['Followers'] = instagram_data['Followers'].apply(value_to_float)
instagram_data['Authentic engagement\r\n'] = instagram_data['Authentic engagement\r\n'].apply(value_to_float)
instagram_data['Engagement avg\r\n'] = instagram_data['Engagement avg\r\n'].apply(value_to_float)
```

```python
#"Engagement" basically is based on influencers' followers, likes/post, and comments/post.

#On the other hand, "Authentic Engagement metric shows the number of likes and comments that on average come from 'Real people' and Influencers per post." (https://help.hypeauditor.com/en/articles/3448596-what-is-authentic-engagement)

#When fake and spam accounts, bots, are involved, large engagement numbers are likely to bring false impressions and a lack of authenticity behind the account. 

#"it is imperative to decipher between authentic activity and paid-for bot account activity."(https://blog.pulse-advertising.com/instagram-bots-authentic-vs-automated-engagement)
```

```python
#So take a look at the columns "Authentic engagement\r\n" and "Engagement avg\r\n"

instagram_data[["Authentic engagement\r\n", "Engagement avg\r\n"] ]
```

```python
# add "AE&E__ratio" to see the Authentic engagement percentage among total Engagement avg

instagram_data["AE&E__ratio"] = instagram_data.apply(lambda row: row["Authentic engagement\r\n"] / row["Engagement avg\r\n"], axis=1)
instagram_data                                                                                      
```

```python
instagram_data.sort_values(by = ["AE&E__ratio"], ascending=False)
```

```python
# Look at the 10% of the top Authentic engagement rate

instagram_data.sort_values(by = ["AE&E__ratio"], ascending=False).head(100)
```

```python
#As we assume - The higher "AE&E__ratio" means a higher genuine interaction taking place with this influencer and account
#Look at the account which AE&E__ratio = 1(highest)

instagram_data.loc[instagram_data["AE&E__ratio"]==1]
```
